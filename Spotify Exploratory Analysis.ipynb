<h3>INFORMATION ON DIFFERENT TERMS USED</h3>

**Acousticness** - 
This value describes how "acoustic" the song is.
Value above 0.66: The song mostly (or completely) contains words
Value between 0.33 and 0.66: The song contains both words and music
Value below 0.33: The song mostly (or completely) contains music
                
**Danceability** - 
How suitable a song is for dancing. 
Based on a combination of musical elements including tempo, rhythm, stability, beat strength and overall   regularity

**Energy** - 
Perceptual measure of intensity and activity.

**Instrumentalness** -
Predicts whether the song contains no vocals.

**Liveness** -
Whether the song was done live or in a studio. Higher the value, greater the probability of the song recorded live.

**Loudness** -
Overall loudness of a song in decibel.

**Valence** -
Describes the musical positivess conveyed by the song. 
Higher the valence, greater the positiveness in the songs.

**Tempo** -
Overall tempo of the song in bpm.

**Speechiness** -
Detects the presence of vocals in a track.

**Key**-
The estimated overall key of the track. Integers map to pitches using standard Pitch Class otation. E.g. 0 = C, 1 = C‚ôØ/D‚ô≠, 2 = D, and so on.

**Mode** -
Indicates modality (major or minor) of the song. Major is 1 and minor is 0.

**Time Signature** -
An estimated overall time signature of a track. The time signature (meter) is a notational convention to specify how many beats are in each bar (or measure).

<h4>How did different features of the songs change with years?</h4>

```python
plt.figure(figsize=(14,7))
sns.set(style="whitegrid")
columns = ['acousticness','instrumentalness','danceability','energy','speechiness','valence']
for col in columns:
    x = track_data.groupby("year")[col].mean()
    ax= sns.lineplot(x=x.index,y=x,label=col)
ax.set_title('How did different song features changed over the years?')
ax.set_ylabel('Feature values')
ax.set_xlabel('Year')
plt.legend(loc=1, frameon=False)
```

```python
plt.figure(figsize=(14,7))
sns.set(style="whitegrid")
columns = ['loudness']
for col in columns:
    x = track_data.groupby("year")[col].mean()
    ax= sns.lineplot(x=x.index,y=x)
ax.set_title('How did loudness changed over the years?')
ax.set_ylabel('Loudness value')
ax.set_xlabel('Year')
```

<h3><u>Explicitness of the tracks</u></h3>

```python
k=0
l=0
exp=[]
for i in track_data['explicit']:
    if i==1:
        k=k+1
    else:
        l=l+1
exp.append(k)
exp.append(l)

y=np.array(exp)
ylabels = ['explicit','non-explicit']
yexplode=[0.2,0]
ycolors=['#D0114D','#D5527D']

plt.figure()
plt.pie(y,labels=ylabels,colors=ycolors,explode=yexplode,shadow=True)
plt.show()
```

<h4>How do the features of the track differentiate as according to their explicit-ness?</h4>

```python
pysql('SELECT explicit, AVG(acousticness), AVG(loudness), AVG(instrumentalness), AVG(danceability), AVG(speechiness), AVG(popularity), AVG(mode), AVG(valence) FROM track_data WHERE explicit=0 UNION SELECT explicit, AVG(acousticness), AVG(loudness), AVG(instrumentalness), AVG(danceability), AVG(speechiness), AVG(popularity), AVG(mode), AVG(valence) FROM track_data WHERE explicit=1')
```

<h3>Top 20 artists according to nummber of songs they featured in</h3>

```python
l = []
for i in track_data.artists:
    a=i[2:-2]
    l.append(a)
l = list(set(l))
lst = []
for i in l:
    res = i.strip('][').split(', ')
    lst.append(res)
d=[]
for i in lst:
    for j in i:
        d.append(j)
dff=pd.DataFrame(d)
dff = dff[0].value_counts()

dff = dff[:20, ]
plt.figure(figsize = (9.5, 6))
sns.barplot(dff.values, dff.index, alpha = 0.9,palette=("spring_d"))
plt.title("Top 20 Artists")
plt.ylabel("Artists")
plt.xlabel("Number of tracks")
plt.show()
plt.subplots_adjust(left=0.27)
```

<h2>CONCLUSION</h2>

1. In figure-3, we can see that there are strong correlation between energy & loudness (0.78), danceability & valence (0.54) and popularity & year (0.51). We can infer that recency bias is affecting the popularity values.
2. In figure-4, we can see that over time acousticness of the tracks has decreased significantly. Other than acousticness, energy of tracks has seen a significant change over time, becoming more energetic. Rest of the features of tracks have not seen significant change.
    From the period of 1920 to around 1950, owes to the lesser number of track records as compared to recent times.
3. From the 'explicit-ness of tracks', we can see that on an average explicit tracks are significantly lesser _acoustic_ and contains more vocal than the non-explicit songs and explicit songs are also more popular than the non-explicit tracks. But, on an average, both explicit and non-explicit tracks have similar positivity score (valence).
